```python
import pandas as pd
import numpy as np
df = pd.read_csv('Titanic.csv')
print(df)
```

```python
df = pd.read_csv('Titanic.csv', on_bad_lines='skip')
print(df)
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop(['Cabin'], axis=1, inplace=True, errors='ignore')

print(df.head())


```

```python
df.drop(['PassengerId' ,'Name','Ticket',],axis=1,inplace=True , errors='ignore')
df['Sex']= df['Sex'].map({'male':0,'female':1})
print(df.head())
```

```python
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True, errors='ignore')
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python

sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Sex vs Survival")
plt.show
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Pclass VS Survival")
plt.show
```

```python
# Age Distribution 
sns.histplot(data=df,x='Age',hue='Survived',multiple='stack')
plt.title('Age Distribution with Survival')
plt.show()
```

```python
# Correlation Matrix 
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Correlation matrix
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Outliers Detection

sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
plt.show()
```
